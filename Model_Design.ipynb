## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
type_counts_df = pd.DataFrame(application_df['APPLICATION_TYPE'].value_counts(),columns = ['count'])
types_to_replace = []

for type_ in type_counts_df.index:
    if type_counts_df.loc[type_,'count'] < 500:
        types_to_replace.append(type_)

# Replace in dataframe
for app in types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts_df = pd.DataFrame(application_df['CLASSIFICATION'].value_counts(),columns=['count'])
class_counts_df[class_counts_df['count'] > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classes_to_replace = []

for class_ in class_counts_df.index:
    if class_counts_df.loc[class_,'count'] < 1000:
        classes_to_replace.append(class_)

# Replace in dataframe
for cls in classes_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
for column in application_df.columns:
    print(application_df[column].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# It would seem wiser to take the 'other' step on the other columns as well.
application_df_dummies = pd.get_dummies(application_df,columns=[
    'APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','INCOME_AMT','SPECIAL_CONSIDERATIONS'
])
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_dummies.drop(columns='IS_SUCCESSFUL')
y = application_df_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80,activation='relu',input_dim=len(X.columns)))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30,activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','precision','recall'])
```

```python
# Train the model
nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy, model_precision, model_recall = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}\nPrecision: {model_precision}, Recall: {model_recall}")
```

Those aren't good results. Next, I'll test a simpler neural network. 80 and 30 perceptrons in the layers seems like overkill.

```python
# Define the second model

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(tf.keras.layers.Dense(units=5,activation='relu',input_dim=len(X.columns)))

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=3,activation='relu'))

# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# Check the structure of the model
nn2.summary()
```

```python
# Compile the model
nn2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','precision','recall'])

# Train the model
nn2.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model2_loss, model2_accuracy, model2_precision, model2_recall = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model2_loss}, Accuracy: {model2_accuracy}\nPrecision: {model2_precision}, Recall: {model2_recall}")
```

```python
# For ease of comparison, the stats from the first model:
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}\nPrecision: {model_precision}, Recall: {model_recall}")
```

The second, simpler model performs more poorly, but only slightly. There may be a better approach than adding more perceptrons. Next, let's try a model with three hidden layers.

```python
# Define the third model

nn3 = tf.keras.models.Sequential()

# First hidden layer
nn3.add(tf.keras.layers.Dense(units=16,activation='relu',input_dim=len(X.columns)))

# Second hidden layer
nn3.add(tf.keras.layers.Dense(units=8,activation='relu'))

# Third hidden layer
nn3.add(tf.keras.layers.Dense(units=4,activation='relu'))

# Output layer
nn3.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

nn3.summary()
```

```python
# Compile the model
nn3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','precision','recall'])

# Train the model
nn3.fit(X_train_scaled,y_train,epochs=100)
```

```python
model3_loss, model3_accuracy, model3_precision, model3_recall = nn3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model3_loss}, Accuracy: {model3_accuracy}\nPrecision: {model3_precision}, Recall: {model3_recall}")
```

Adding a third layer did not substantially improve the results. Let's try different activation functions.

```python
# Define the fourth model

nn4 = tf.keras.models.Sequential()

# First hidden layer
nn4.add(tf.keras.layers.Dense(units=80,activation='linear',input_dim=len(X.columns)))

# Second hidden layer
nn4.add(tf.keras.layers.Dense(units=40,activation='linear'))

# Third hidden layer
nn4.add(tf.keras.layers.Dense(units=20,activation='linear'))

# Output layer
nn4.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

nn4.summary()
```

```python
# Compile the model
nn4.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy','precision','recall'])

# Train the model
nn4.fit(X_train_scaled,y_train,epochs=100)
```

```python
model4_loss, model4_accuracy, model4_precision, model4_recall = nn4.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model4_loss}, Accuracy: {model4_accuracy}\nPrecision: {model4_precision}, Recall: {model4_recall}")
```

Still no improvement. The models all reach their maximum accuracy very quickly, which suggests that a different approach, rather than a bigger approach would be best. It's possible that the data provided don't determine the success or failure of the campaigns. Assuming that that's not the case, continuing to tweak the activation functions would be the best next step.
